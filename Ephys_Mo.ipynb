```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pyabf
import seaborn as sns

print("We're using pyabf v{}".format(pyabf.__version__))
```

```python
abf = pyabf.ABF("21506009.abf")
print(abf)
abf.setSweep(4)
print("sweep data (ADC):", abf.sweepY)
print("sweep command (DAC):", abf.sweepC)
print("sweep times (seconds):", abf.sweepX)
plt.figure(figsize=(8, 5))
plt.plot(abf.sweepX, abf.sweepY)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# plt.title("pyABF and Matplotlib are a great pair!")
plt.ylabel(abf.sweepLabelY)
plt.xlabel(abf.sweepLabelX)
for i in [0, 1,5]:
    abf.setSweep(i)
    plt.plot(abf.sweepX, abf.sweepY, alpha=1, label="sweep %d" % (i))
# plt.legend('')
plt.xlim(0,1.5)
plt.ylim(-100,40)
plt.savefig('RFPexample_action_pontentials_scale_bar_21506009_abf',bbox_inches='tight',dpi=150,transparent=True)

# plt.show()
```

```python
plt.figure(figsize=(8, 5))
# plt.title("pyABF and Matplotlib are a great pair!")
plt.ylabel(abf.sweepLabelY)
plt.xlabel(abf.sweepLabelX)
for i in [0, 1,5]:
    abf.setSweep(i)
    plt.plot(abf.sweepX, abf.sweepY, alpha=1, label="sweep %d" % (i))
# plt.legend('')
plt.axis('off')
plt.xlim(0,1.5)
plt.ylim(-100,40)
plt.savefig('RFPexample_action_pontentials_21506009_abf',bbox_inches='tight',dpi=150,transparent=True)

# plt.show()
```
